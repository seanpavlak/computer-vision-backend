```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
```

```python
def dct_basis_function(basis_count):
    basis_array = np.zeros((basis_count, basis_count))
    
    for p in range(basis_count):
        for q in range(basis_count):
            if (p == 0):
                basis_array[p, q] = np.sqrt(1/basis_count)
            else:
                basis_array[p, q] = np.sqrt(2/basis_count)*np.cos((np.pi*(2*q+1)*p)/(2*basis_count))
    
    basis_array_flat = basis_array.flatten()
    basis_array_flat.shape = (basis_array_flat.shape[0], 1)
    dct_basis_function = np.transpose(basis_array_flat)*basis_array_flat
        
    scipy.misc.imsave('E1_dct_basis_function_%dx%d.png' %(basis_count, basis_count), dct_basis_function)
    
    plt.figure()
    plt.imshow(dct_basis_function, 'gray')
    plt.xticks([]), plt.yticks([])
    plt.title('Basis Function %dx%d' %(basis_count, basis_count))
    plt.show()
    
    return dct_basis_function
```

```python
dct_basis_function_4x4 = dct_basis_function(4)
```
